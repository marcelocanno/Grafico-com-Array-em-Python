```python
import igraph as ig
import matplotlib.pyplot as plt
```

```python
g = ig.Graph(
    6,
    [(0, 1), (0, 2), (1, 3), (2, 3), (2, 4), (3, 5), (4, 5)]
)
```

```python
g.es['width'] = 0.5
```

```python
fig, ax = plt.subplots()
ig.plot(
    g,
    target=ax,
    layout='circle',
    vertex_color='steelblue',
    vertex_label=range(g.vcount()),
    edge_width=g.es['width'],
    edge_color='#666666',
    edge_background='white'
)
plt.show()
```
